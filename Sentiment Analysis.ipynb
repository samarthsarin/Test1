```python
from nltk.tokenize import word_tokenize
import pandas as pd
from tqdm import tqdm
import string
from nltk.corpus import stopwords
from string import punctuation
import warnings
warnings.filterwarnings('ignore')
import re
tqdm.pandas()
```

```python
with open('C:/Users/samarth152721/Desktop/negative-words.txt') as f:
    words = f.readlines()    
clean_words = [word[:-1] for word in words]
```

```python
df = pd.read_csv('Clean_Complaints_Analysis.csv')
```

```python
def scoring(text):
    count = 0
    words = word_tokenize(text)
    for w  in words:
        if w in clean_words:
            count+=1   
    return count/len(words)
```

```python
df1 = df[:10000]
df['Consumer complaint narrative'] = df['Consumer complaint narrative'].str.lower()
df.head()
```

```python
df1['Sentiment Score'] = df1['Consumer complaint narrative'].progress_apply(scoring)
```

```python
%%time
from nltk.sentiment.vader import SentimentIntensityAnalyzer
senti = SentimentIntensityAnalyzer()
score = []
for i in tqdm(range(len(df1))):
    score.append(senti.polarity_scores(df1['Consumer complaint narrative'][i])['neg'])
```

```python
df1['Vader Sentiment Score'] = score
```

```python
df1.sort_values(by=['Vader Sentiment Score'],ascending=False).head(50)
```

```python
def getting_amount(text):
    amount = 0.0
    currency = re.findall('\d+\.\d+',text)
    curr = [float(i) for i in currency]
    try:
        amount = max(curr)
    except:
        pass
    return amount
```

```python
df1['Amount'] = df1['Consumer complaint narrative'].progress_apply(getting_amount)
```

```python
df1.head(100)
```

```python
df1['Total Sentiment Score'] = ''
for i in tqdm(range(len(df1))):
    df1['Total Sentiment Score'][i] = 0.5*df1['Sentiment Score'][i]+0.5*df1['Vader Sentiment Score'][i]
    

df1['Total Score'] = ''
for i in tqdm(range(len(df1))):
    df1['Total Score'][i] = 0.4*df1['Sentiment Score'][i]+0.4*df1['Vader Sentiment Score'][i]+0.2*df1['Amount'][i]
```
