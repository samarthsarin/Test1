```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import style
import seaborn as sns
%matplotlib inline
import numpy as np
```

```python
df = pd.read_csv('C:/Users/samarth152721/Downloads/complaints.csv')
df[df['Product']=='Credit card or prepaid card']
```

```python
df['Date received'] = pd.to_datetime(df['Date received'])
df['Date sent to company'] = pd.to_datetime(df['Date sent to company'])
```

```python
date = df.groupby(df['Date received'].dt.year)
```

```python
date.head()
```

```python
dates = df['Date received'].groupby([df['Date received'].dt.year, df['Date received'].dt.month]).agg('count')
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None): 
    print(dates)
```

# Overall

```python
plt.figure(figsize=(20,12))
dates.plot(kind= 'bar')
plt.show()
```

# Year Wise

```python
year_wise = df['Date received'].groupby(df['Date received'].dt.year).agg('count')
plt.figure(figsize=(20,12))
print(year_wise.head(10))
year_wise.plot(kind= 'bar')
plt.show()
```

# Month Wise

```python
month_wise = df['Date received'].groupby(df['Date received'].dt.month).agg('count')
plt.figure(figsize=(20,12))
print(month_wise.head(12))
month_wise.plot(kind= 'bar',color='green')
plt.show()
```
